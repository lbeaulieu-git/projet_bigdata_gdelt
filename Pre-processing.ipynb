## Scrapping

##### masterlielist.txt

```python
from io import BytesIO
from zipfile import ZipFile
from urllib.request import urlopen
import requests

def scrapper_url_zip(url = "http://data.gdeltproject.org/gdeltv2/masterfilelist.txt"):

    response = requests.get(url)
    content = response.content.decode("utf-8") 
    l = content.split('\n')[:9]

    liste = list()
    list_df = list()

    for i in l:
        k = i.split(" ")[-1]
        liste.append(k)

    for i in liste:
        url = urlopen(i)

        k = i.split("/")[-1].split(".zip")[0]

        zipfile = ZipFile(BytesIO(url.read()))
        FFdata = pd.read_csv(zipfile.open(k), header=None,on_bad_lines='skip', sep="\t")
        list_df.append(FFdata)
    return list_df
    
df_base = scrapper_url_zip()
df_base[0]
```

### masterfilelist-translation.txt

```python
def scrapper_url_zip_translation(url = "http://data.gdeltproject.org/gdeltv2/masterfilelist-translation.txt"):

    response = requests.get(url)
    content = response.content.decode("utf-8") 
    l = content.split('\n')[:9]

    liste = list()
    list_df = list()

    for i in l:
        k = i.split(" ")[-1]
        liste.append(k)

    for i in liste:
        url = urlopen(i)

        k = i.split("/")[-1].split(".zip")[0]

        zipfile = ZipFile(BytesIO(url.read()))
        FFdata = pd.read_csv(zipfile.open(k), header=None,on_bad_lines='skip', sep="\t")
        list_df.append(FFdata)
    return list_df
    
df_translation = scrapper_url_zip_translation()
df_translation[0]
```

### Pre-processing base table

```python
def sub_df_base(df_list):
    
    list_df_export = list()
    list_df_mentions = list()
    list_df_gkg = list()
    
    for df in df_list:
        if df.shape[1] == 61:
            #export
            list_df_export.append(df)
            
    for df in df_list:
        if df.shape[1] == 16:
            #mention
            list_df_mentions.append(df)
            
    for df in df_list:
        if df.shape[1] == 27:
            #gkg
            list_df_gkg.append(df)
            
    df_export_base = pd.concat(list_df_export)
    df_mentions_base = pd.concat(list_df_mentions)    
    df_gkg_base = pd.concat(list_df_gkg)
    
    for i in range(df_export_base.shape[1]):
        df_export_base.rename({i: 'export_'+str(i)}, axis=1, inplace=True)
        
    for i in range(df_mentions_base.shape[1]):
        df_mentions_base.rename({i: 'mentions_'+str(i)}, axis=1, inplace=True)    
        
    for i in range(df_gkg_base.shape[1]):
        df_gkg_base.rename({i: 'gkg_'+str(i)}, axis=1, inplace=True)    
    
    return df_export_base, df_mentions_base, df_gkg_base

export, mentions, gkg = sub_df_base(df_base)

export
```

### Pre-processing translation table

```python
def sub_df_base_translation(df_list_translation):
    
    list_df_export_translation = list()
    list_df_mentions_translation = list()
    list_df_gkg_translation = list()
    
    for df in df_list_translation:
        if df.shape[1] == 61:
            #export
            list_df_export_translation.append(df)
            
    for df in df_list_translation:
        if df.shape[1] == 16:
            #mention
            list_df_mentions_translation.append(df)
            
    for df in df_list_translation:
        if df.shape[1] == 27:
            #gkg
            list_df_gkg_translation.append(df)
            
    df_export_base_translation = pd.concat(list_df_export_translation)
    df_mentions_base_translation = pd.concat(list_df_mentions_translation)    
    df_gkg_base_translation = pd.concat(list_df_gkg_translation)
    
    for i in range(df_export_base_translation.shape[1]):
        df_export_base_translation.rename({i: 'export_translation_'+str(i)}, axis=1, inplace=True)
        
    for i in range(df_mentions_base_translation.shape[1]):
        df_mentions_base_translation.rename({i: 'mentions_translation_'+str(i)}, axis=1, inplace=True)    
        
    for i in range(df_gkg_base_translation.shape[1]):
        df_gkg_base_translation.rename({i: 'gkg_translation_'+str(i)}, axis=1, inplace=True)    
    
    return df_export_base_translation, df_mentions_base_translation, df_gkg_base_translation

export_translation, mentions_translation, gkg_translation = sub_df_base_translation(df_translation)

export_translation
```

### Pour un pays donnÃ© en paramÃ¨tre, affichez les Ã©vÃ¨nements qui y ont eu place triÃ©es par le nombre de mentions (tri dÃ©croissant); permettez une agrÃ©gation par jour/mois/annÃ©e

```python
def display_event(event_table, country, day, month, year):
    event_table[1] = pd.to_datetime(event_table["export_1"], format='%Y%m%d') # column_1 = Day 
    list_event = event_table.loc[event_table["export_53"] == country]
    list_event_sorted = list_event.sort_values(by="export_31") # column_31 = NumMentions
    #print(list_event_sorted)
    list_event_sorted['day'] = list_event_sorted[1].dt.day
    list_event_sorted['month'] = list_event_sorted[1].dt.month
    list_event_sorted['year'] = list_event_sorted[1].dt.year
    
    event_by_day_month_year = list_event_sorted.loc[(list_event_sorted.day == day) & (list_event_sorted.month == month) & (list_event_sorted.year == year)]
    event_by_month_year = list_event_sorted.loc[(list_event_sorted.month == month) & (list_event_sorted.year == year)]
    event_by_year = list_event_sorted.loc[list_event_sorted.year == year]
    #print(event_by_year)
    print("--- Events in " + str(day) + "/" + str(month) + "/" + str(year) +" ---\n")
    if event_by_day_month_year["export_0"].empty:
        print("0 events found")
    else:
        print(event_by_day_month_year["export_0"])
    print("\n\n--- Events in " + str(month) + "/" + str(year) +" ---\n")
    if event_by_month_year["export_0"].empty:
        print("0 events found")
    else:
        print(event_by_month_year["export_0"])
    print("\n\n--- Events in " + str(year) + " ---\n")
    if event_by_year["export_0"].empty:
        print("0 events found")
    else:
        print(event_by_year["export_0"])

display_event(export, "SW", 19, 1, 2015)
```

### Pour une source de donnÃ©s passÃ©e en paramÃ¨tre (gkg.SourceCommonName) affichez les thÃ¨mes, personnes, lieux dont les articles de cette sources parlent ainsi que le nombre dâ€™articles et le ton moyen des articles (pour chaque thÃ¨me/personne/lieu); permettez une agrÃ©gation par jour/mois/annÃ©e.

```python
from statistics import mean

def source_stats(gkg_table, source, day=None, month=None, year=None):
#     # column_3 = SourceCommonName
#     # column_7 = Theme
#     # column_11 = Person
#     # column_9 = Locations
#     # column_15 = Avg_Tone
#     gkg_table[1] = pd.to_datetime(gkg_table[1], format='%Y%m%d%H%M%S') # column_1 = Day 
#     list_event = gkg_table.loc[event_table[3] == source]
#     list_theme = list_event.loc(list_event)
    

    themes = list()
    persons = list()
    locations = list()

    df_avg_tone_theme = dict()
    df_avg_tone_person = dict()
    df_avg_tone_location = dict()

    #gkg_table["gkg_1"] = pd.to_datetime(gkg_table["gkg_1"], format='%Y%m%d%H%M%S') # column_1 = Day 
    list_event = gkg_table.loc[gkg_table["gkg_3"] == source]
    list_theme = list_event["gkg_7"].tolist()
    list_person = list_event["gkg_11"].tolist()
    list_location = list_event["gkg_9"].tolist()

    # THEME
    list_theme = [t for t in list_theme if str(t) != 'nan'] # drop nan values
    for theme in list_theme:
        theme_temp_list = str(theme).split(";")
        theme_temp_list.pop() # drop the last element of the list which is an empty string
        for theme_sub in theme_temp_list:
            themes.append(theme_sub)
    themes = list(dict.fromkeys(themes)) # drop duplicated themes

    # PERSON
    list_person = [l for l in list_person if str(l) != 'nan'] # drop nan values
    for person in list_person:
        person_temp_list = str(person).split(";")
        for person_sub in person_temp_list:
            persons.append(person_sub)
    persons = list(dict.fromkeys(persons)) # drop duplicated themes

    # LOCATION
    list_location = [l for l in list_location if str(l) != 'nan'] # drop nan values
    for location in list_location:
        location_temp_list = str(location).split("#")
        locations.append(location_temp_list[1])
    locations = list(dict.fromkeys(locations)) # drop duplicated themes
    
    print("---- SOURCE = " + source + "----\n")
    
    print("\n---- THEMES ----\n")
    print(themes)
    print("\n\n---- PERSONS ----\n")
    print(persons)
    print("\n\n---- LOCATIONS ----\n")
    print(locations)

    # TONE AVG PER THEMES
    for theme in themes:
        list_theme_ = list_event["gkg_7"].astype(str)
        list_theme_ = list_theme_[list_theme_.str.contains(theme)]
        idx_theme = list_theme_.index.tolist()
        avg_tone_theme = list()
        for i in idx_theme:
            avg_temp = gkg.iloc[i,15].split(",")[0]
            avg_tone_theme.append(float(avg_temp))

        avg_tone_theme = mean(avg_tone_theme)
        df_avg_tone_theme[theme] = avg_tone_theme
    df_avg_tone_theme = pd.DataFrame(df_avg_tone_theme.items(), columns=['Themes', 'AVG_Tone'])

    # TONE AVG PER PERSON
    for person in persons:
        list_person_ = list_event["gkg_11"].astype(str)
        list_person_ = list_person_[list_person_.str.contains(person)]
        idx_person = list_person_.index.tolist()
        avg_tone_person = list()
        for i in idx_person:
            avg_temp_person = gkg.iloc[i,15].split(",")[0]
            avg_tone_person.append(float(avg_temp_person))

        avg_tone_person = mean(avg_tone_person)
        df_avg_tone_person[person] = avg_tone_person
    df_avg_tone_person = pd.DataFrame(df_avg_tone_person.items(), columns=['Persons', 'AVG_Tone'])
    df_avg_tone_person

    # TONE AVG PER LOCATION
    for location in locations:
        list_location_ = list_event["gkg_9"].astype(str)
        list_location_ = list_location_[list_location_.str.contains(location)]
        idx_location = list_location_.index.tolist()
        avg_tone_location = list()
        for i in idx_location:
            avg_temp_location = gkg.iloc[i,15].split(",")[0]
            avg_tone_location.append(float(avg_temp_location))

        avg_tone_location = mean(avg_tone_location)
        df_avg_tone_location[location] = avg_tone_location
    df_avg_tone_location = pd.DataFrame(df_avg_tone_location.items(), columns=['Locations', 'AVG_Tone'])
    
    print("\n\n\n\n---- AVG TONE PER THEMES ----\n")
    print(df_avg_tone_theme)
    print("\n\n---- AVG TONE PER PERSONS ----\n")
    print(df_avg_tone_person)
    print("\n\n---- AVG TONE PER LOCATIONS ----\n")
    print(df_avg_tone_location)
    
source_stats(gkg, "wjon.com")
```
